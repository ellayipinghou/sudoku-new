```python
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping
import datetime
import numpy as np
```

```python
from tensorflow.keras.layers import Input
```

```python
import os
import cv2
```

```python
X = []
y = []
for i in range(10):
    for d in os.listdir("../assets/{}".format(i)):
        t_img = cv2.imread("../assets/{}".format(i)+"/"+d)
        t_img = cv2.cvtColor(t_img,cv2.COLOR_BGR2GRAY)
        X.append(t_img)
        y.append(i)
```

```python
X = np.array(X)
y = np.array(y)
```

```python
display(X.shape)
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test, y_train, y_test = train_test_split(X,y,test_size = 0.20, random_state= 21)
```

```python
X_train.shape
```

```python
y_test
```

```python
X_train = X_train.reshape((X_train.shape[0], 28, 28, 1)).astype('float32')
X_test = X_test.reshape((X_test.shape[0], 28, 28, 1)).astype('float32')
# normalize inputs from 0-255 to 0-1
X_train = X_train / 255
X_test = X_test / 255
# one hot encode outputs
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
num_classes = y_test.shape[1]
```

```python
num_classes
```

```python
def larger_model():
    # create model
	model = keras.Sequential(
    [
        Input(shape=(28, 28, 1)),
        Conv2D(30, (5, 5), input_shape=(28, 28, 1), activation='relu'),
        MaxPooling2D(),
        Conv2D(15, (3, 3), activation='relu'),
        MaxPooling2D(),
        Dropout(0.2),
        Flatten(),
        Dense(128, activation='relu'),
        Dense(50, activation='relu'),
        Dense(num_classes, activation='softmax')
    ])
    

    # Compile model
	model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
	return model
```

```python
model = larger_model()
log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = keras.callbacks.TensorBoard(log_dir=log_dir, histogram_freq=1)
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)
```

```python
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=20, batch_size=10,callbacks=[tensorboard_callback, early_stopping])
# Final evaluation of the model
scores = model.evaluate(X_test, y_test, verbose=0)

print("Large CNN Error: %.2f%%" % (100-scores[1]*100))
```

```python
scores
```

```python
pred = model.predict(X_test,verbose = 1)
pred_classes = np.argmax(pred, axis=1)
```
